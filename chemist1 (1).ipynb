```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = pd.DataFrame({
    'Medicine': ['A', 'B', 'C', 'D'],
    'Stock Level': [100, 200, 50, 80]
})

plt.figure(figsize=(10,6))
plt.bar(data['Medicine'], data['Stock Level'], color='skyblue')
plt.title('Medicine Stock Level')
plt.xlabel('Medicine')
plt.ylabel('Stock Level')
plt.show()
```

```python
# Assuming you have a DataFrame with daily sales data for a particular medicine
data = pd.DataFrame({
    'Date': pd.date_range(start='2023-01-01', periods=7, freq='D'),
    'Sales': [20, 35, 40, 55, 60, 50, 70]
})

plt.figure(figsize=(10,6))
plt.plot(data['Date'], data['Sales'], marker='o', color='green')
plt.title('Sales Trend Over the Last Week')
plt.xlabel('Date')
plt.ylabel('Units Sold')
plt.xticks(rotation=45)
plt.show()
```

```python
data = pd.DataFrame({
    'Medicine': ['A', 'B', 'C', 'D'],
    'Units Sold': [100, 150, 80, 50]
})

plt.figure(figsize=(10,6))
plt.bar(data['Medicine'], data['Units Sold'], color='orange')
plt.title('Top-Selling Medicines')
plt.xlabel('Medicine')
plt.ylabel('Units Sold')
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(data['Units Sold'], labels=data['Medicine'], autopct='%1.1f%%', startangle=90)
plt.title('Market Share of Medicines Sold')
plt.show()
```
